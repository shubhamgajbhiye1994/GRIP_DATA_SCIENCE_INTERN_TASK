```
# histogram plot

def hist_plt(df = data):
  plt.figure(figsize=(15,15))
  i = 1
  for col in df.columns:
    if df[col].dtype != 'object':
      plt.subplot(3,3,i)
      sb.histplot(df[col],bins=100)
      i += 1
```

```
hist_plt()
```

```
# Density plot for distribution of continuous variable

def kde_plt(df = data):
  plt.figure(figsize=(15,15))
  i = 1
  for col in df.columns:
    if df[col].dtype != 'object':
      plt.subplot(3,3,i)
      sb.kdeplot(df[col],)
      i += 1
```

```
kde_plt()
```

```
# plotting box for outlier and densirty check

def box_plt(df = data):
  plt.figure(figsize=(15,15))
  i = 1
  for col in df.columns:
    if df[col].dtype != 'object':
      plt.subplot(3,3,i)
      sb.boxplot(df[col])
      i += 1
```

```
box_plt()
```

***Categorical variable analysis***

```
# plotting count of categorical variables

def count_plt(df = data):
  plt.figure(figsize=(50,50))
  i = 1
  for col in df.columns:
    if df[col].dtype == 'object':
      plt.subplot(3,3,i)
      sb.countplot(df[col])
      i += 1
```

```
count_plt()
```

```
# Plotting categorical with continious values i.e. Level 1 

def categ_vs_continu_lvl_1(group_val,df = data , sum = True,size = (100,100) ):

  if sum == True:
    temp = df.groupby(by=group_val).sum()
  else:
    temp = df.groupby(by=group_val)
  i = 1
  temp.reset_index(inplace=True)
  plt.figure(figsize=size)
  for col in temp.columns:
    if col not in group_val:
      plt.subplot(len(temp.columns),2,i)
      sb.barplot(y=col,x = group_val ,data=temp)
      i += 1
```

```
categ_vs_continu_lvl_1('Ship Mode',size = (20,20))
```

```
categ_vs_continu_lvl_1('Segment',size = (20,20))
```

```
categ_vs_continu_lvl_1('Region',size = (20,20))
```

```
categ_vs_continu_lvl_1('Sub-Category',size = (50,50))
```

```
categ_vs_continu_lvl_1('Category',size = (20,20))
```
